```python
budzet = 50
cena = 65
if cena <= budzet:                       # warunek + DWUKROPEK
    print("Kupuję!")                     # gałąź na True — WCIĘTA
else:                                    # else też z dwukropkiem
    print(f"Brakuje {cena - budzet} zł") # gałąź na False — WCIĘTA
print("Koniec zakupów")                  # bez wcięcia — wykona się ZAWSZE
```
